```python
#---------------------#
#   Define function   #
#---------------------#

def mtf(a, c, d):
    m = a + c[0]*d[0] + c[1]*d[1] + c[2]*d[2] +c[3]*d[3]
    return m 


#------------------#
#   Coefficients   #
#------------------#

#--- All data
G0_a            = [2.4,1.57]   # count, cmass
G0_a_CI         = [0.972,1.817]   # lower, upper; only count
G0_a_ste        = 0.13

G0_c_dbh        = [ 0.04, 0.04]
G0_c_dbh_CI     = [ 0.031,0.045]
G0_c_dbh_ste    = 0.004

G0_c_conifer     = [ 0,0.45]
G0_c_conifer_CI  = [ 0.131,0.691]
G0_c_conifer_ste = 0

G0_c_aridity     = [ 0,0.42]
G0_c_aridity_CI  = [ 0.210, 0.516]
G0_c_aridity_ste = 0.

G0_c_MAT        = [-0.1238,-0.13]
G0_c_MAT_CI     = [-0.157,-0.090]
G0_c_MAT_ste    = 0.017


# --- Fire coefficients
G1_a             = [1.51,2.71]
G1_a_CI          = [-0.307,0.813]
G1_a_ste         = 0.22

G1_c_dbh         = [0.05,2.71]
G1_c_dbh_CI      = [0.029,0.046]
G1_c_dbh_ste     = 0.0

G1_c_conifer     = [0,0]
G1_c_conifer_CI  = [0,0]
G1_c_conifer_ste = 0

G1_c_aridity     = [0,0]
G1_c_aridity_CI  = [0,0]
G1_c_aridity_ste = 0

G1_c_MAT         = [0,0]
G1_c_MAT_CI      = [0,0]
G1_c_MAT_ste     = 0

#--- No fire coefficients
G2_a             = [ 2.51,2.80]
G2_a_CI          = [2.181,2.831]
G2_a_ste         = 0.163

G2_c_dbh         = [ 0.02,0.02]
G2_c_dbh_CI      = [0.006,0.032]
G2_c_dbh_ste     = 0.006

G2_c_conifer     = [ 0.39,0.4]
G2_c_conifer_CI  = [0.112,0.659]
G2_c_conifer_ste = 0.137

G2_c_aridity     = [ 0,0]
G2_c_aridity_CI  = [ 0,0]
G2_c_aridity_ste = 0

G2_c_MAT         = [-0.12,-0.14]
G2_c_MAT_CI      = [-0.157,-0.091]
G2_c_MAT_ste     = 0.016


#--- Other mort coefficients
G3_a         = [ 2.54,2.81]
G3_a_CI      = [ 2.162,2.845]
G3_a_ste     = 0.17

G3_c_dbh     = [ 0.02,0.02]
G3_c_dbh_CI  = [0.006,0.032]
G3_c_dbh_ste = 0.001

G3_c_MAT     = [-0.14, -0.15]
G3_c_MAT_CI  = [-0.169, -0.095]
G3_c_MAT_ste = 0.02

G3_c_conifer     = [ 0.52,0.49]
G3_c_conifer_CI  = [ 0.2,0.763]
G3_c_conifer_ste = 0.141

G3_c_aridity     = [ 0,0]
G3_c_aridity_CI  = [0,0]
G3_c_aridity_ste = 0


#--- Set coefficients into lists
G0_c      = np.array([G0_c_dbh,G0_c_conifer,G0_c_aridity,G0_c_MAT],dtype='object')
G0_c_CI   = np.array([G0_c_dbh_CI,G0_c_conifer_CI,G0_c_aridity_CI,G0_c_MAT_CI],dtype='object')
G0_c_ste  = np.array([G0_c_dbh_ste,G0_c_conifer_ste,G0_c_aridity_ste,G0_c_MAT_ste],dtype='object')

G1_c      = np.array([G1_c_dbh,G1_c_conifer,G1_c_aridity,G1_c_MAT],dtype='object')
G1_c_CI   = np.array([G1_c_dbh_CI,G1_c_conifer_CI,G1_c_aridity_CI,G1_c_MAT_CI],dtype='object')
G1_c_ste  = np.array([G1_c_dbh_ste,G1_c_conifer_ste,G1_c_aridity_ste,G1_c_MAT_ste],dtype='object')

G2_c      = np.array([G2_c_dbh,G2_c_conifer,G2_c_aridity,G2_c_MAT],dtype='object')
G2_c_CI   = np.array([G2_c_dbh_CI,G2_c_conifer_CI,G2_c_aridity_CI,G2_c_MAT_CI],dtype='object')
G2_c_ste  = np.array([G2_c_dbh_ste,G2_c_conifer_ste,G2_c_aridity_ste,G2_c_MAT_ste],dtype='object')

G3_c      = np.array([G3_c_dbh,G3_c_conifer,G3_c_aridity,G3_c_MAT],dtype='object')
G3_c_CI   = np.array([G3_c_dbh_CI,G3_c_conifer_CI,G3_c_aridity_CI,G3_c_MAT_CI],dtype='object')
G3_c_ste  = np.array([G3_c_dbh_ste,G3_c_conifer_ste,G3_c_aridity_ste,G3_c_MAT_ste],dtype='object')



#--- Plot settings
alpha = 0.3
```
