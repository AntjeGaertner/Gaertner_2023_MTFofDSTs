```python
# Load libraries
import pandas as pd
import pickle
import os

%config Completer.use_jedi = False 

#--- Notebook run settings
run = 'command line' # 'manual' 'command line'
#run = 'manual'


# Settings for the multiple linear regression model (MLR)
#--- Arguments
if run == 'command line':
    # Read command line arguments
    arguments = os.environ['NB_ARGS']
    alpha, date_exp_str_full, date_exp_str_managed, remove_data = arguments.split(',')
    
    alpha              = float(alpha)   
    remove_data        = bool(remove_data)
    
elif run == 'manual':
    # Enter arguments manually 
    alpha                = 0.01
    date_exp_str_full    = '230118_Full'
    date_exp_str_managed = '230118_Management'
    remove_data          = True


with open('spe_reg_results/'+date_exp_str_full+'_best_model_climate_'+str(remove_data)+'_'+str(alpha)+'_species.pkl','rb') as f:
    species_clim_best_models = pickle.load(f)
with open('site_reg_results/'+date_exp_str_managed+'_best_model_climate_'+str(remove_data)+'_'+str(alpha)+'_site.pkl','rb') as f:
    site_clim_best_models = pickle.load(f)    
    
# Finish table    
species_clim_best_models.index = ['Species']*5
species_clim_best_models.index.name = 'Level'
species_clim_best_models = species_clim_best_models.reset_index().set_index(['Level','model'])
site_clim_best_models.index = ['Site']*5
site_clim_best_models.index.name = 'Level'
site_clim_best_models = site_clim_best_models.reset_index().set_index(['Level','model'])

table_climate_comparison = pd.concat([site_clim_best_models,species_clim_best_models],axis=0)
display(table_climate_comparison)

print(table_climate_comparison.to_latex())
```
